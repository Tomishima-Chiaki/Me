```python
def get_disaster_summary(iso_code):
    # ISO
    country_data = emdat[emdat['ISO'] == iso_code]
    
    # 
    country_name = country_data['Country'].iloc[0]
    
    # 
    total_disasters = country_data.shape[0]
    
    # 
    total_deaths = country_data['Total Deaths'].sum()
    
    # 
    print(':', country_name)
    print(':', total_disasters)
    print(':', total_deaths)
```

```python
# 
get_disaster_summary('JPN')

get_disaster_summary('USA')

get_disaster_summary('IND')

get_disaster_summary('CHN')

get_disaster_summary('PHL')
```

```python
# use a loop!
countries = ["JPN","USA","IND","CHN","PHL"]

for country in countries:
    get_disaster_summary(country)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
df = pd.read_csv("emdat.csv")


def plot_deaths_by_disaster(country_name: str):

    """
    
    """
    # 
    country_df = df[df["Country"] == country_name]

    if country_df.empty:
        print(f": {country_name}")
        return
    
    # 
    grouped = country_df.groupby("Disaster Type")["Total Deaths"].sum()

    # 
    plt.figure(figsize=(10, 6))
    grouped.plot(kind="bar")
    plt.xlabel("Disaster Type")
    plt.ylabel("Total Deaths")
    plt.title(f"Total Deaths by Disaster Type in {country_name}")
    plt.tight_layout()
    plt.show()
plot_deaths_by_disaster("Japan")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 
df = pd.read_csv("emdat.csv")


def compare_two_countries(country1: str, country2: str):
    """
    2
    """
    df1 = df[df["Country"] == country1]
    df2 = df[df["Country"] == country2]

    if df1.empty:
        print(f": {country1}")
        return
    if df2.empty:
        print(f": {country2}")
        return
    
    grouped1 = df1.groupby("Disaster Type")["Total Deaths"].sum()
    grouped2 = df2.groupby("Disaster Type")["Total Deaths"].sum()

    # 2
    comparison_df = pd.DataFrame({
        country1: grouped1,
        country2: grouped2
    }).fillna(0)

    # 
    plt.figure(figsize=(12, 6))
    comparison_df.plot(kind="bar")
    plt.xlabel("Disaster Type")
    plt.ylabel("Total Deaths")
    plt.title(f"Total Deaths by Disaster Type\n{country1} vs {country2}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

#  
compare_two_countries("Japan", "Republic of Korea")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("emdat.csv")

def plot_top_deadliest_disasters(n=5):
    df_clean = df.dropna(subset=["Total Deaths"])

    # 
    top_disasters = df_clean.sort_values(by="Total Deaths", ascending=False).head(n)

    # Start Year + Country + Type
    labels = (
        top_disasters["Start Year"].astype(str)
        + " / " +
        top_disasters["Disaster Type"]
        + " (" +
        top_disasters["Country"] + ")"
    )

    plt.figure(figsize=(12, 6))
    plt.bar(labels, top_disasters["Total Deaths"])
    plt.xticks(rotation=45, ha="right")
    plt.xlabel("Disaster (Year / Type / Country)")
    plt.ylabel("Total Deaths")
    plt.title(f"Top {n} Deadliest Disasters")
    plt.tight_layout()
    plt.show()

# 5
plot_top_deadliest_disasters(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_deaths_in_range(df, start_year, end_year, country="Japan"):
    #  + 
    subset = df[
        (df["Country"] == country) &
        (df["Start Year"] >= start_year) &
        (df["Start Year"] <= end_year)
    ].copy()

    # 
    subset = subset.dropna(subset=["Total Deaths"])

    # 
    yearly_deaths = subset.groupby("Start Year")["Total Deaths"].sum()

    # ======  ======
    plt.figure(figsize=(12, 6))
    plt.bar(yearly_deaths.index, yearly_deaths.values)

    plt.xlabel("Year")
    plt.ylabel("Total Deaths")
    plt.title(f"Total Disaster Deaths in {country} ({start_year}{end_year})")

    plt.xticks(yearly_deaths.index, rotation=45)
    plt.tight_layout()
    plt.show()
visualize_deaths_in_range(df, 2000, 2010)
```

```python
import pandas as pd
df = pd.read_csv("emdat.csv")
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

def visualize_asia_storm_earthquake(df):
    # 
    asia_countries = [
        "Japan", "China", "Korea Rep.", "Republic of Korea", "India", "Indonesia",
        "Philippines", "Bangladesh", "Pakistan", "Nepal",
        "Iran (Islamic Republic of)", "Turkey", "Myanmar",
        "Sri Lanka", "Thailand", "Vietnam", "Malaysia", "Mongolia", "Afghanistan"
    ]

    # Disaster Type
    eq_mask = df["Disaster Type"].str.contains("earth", case=False, na=False) | \
              df["Disaster Type"].str.contains("seis", case=False, na=False)

    storm_mask = df["Disaster Type"].str.contains("storm", case=False, na=False)

    asia_se = df[
        (df["Country"].isin(asia_countries)) &
        (eq_mask | storm_mask)
    ].copy()

    if asia_se.empty:
        print(" Storm / Earthquake ")
        return

    counts = asia_se.groupby("Disaster Type").size()
    deaths = asia_se.groupby("Disaster Type")["Total Deaths"].sum()

    bar_colors = ["red" if d >= 1000 else "blue" for d in deaths]

    plt.figure(figsize=(14, 6))

    # (1) 
    plt.subplot(1, 2, 1)
    counts.plot(kind="bar")
    plt.title("Occurrences in Asia (Storm & Earthquake)")
    plt.ylabel("Count")
    plt.xlabel("Disaster Type")

    # (2) 
    plt.subplot(1, 2, 2)
    plt.bar(deaths.index, deaths.values, color=bar_colors)
    plt.title("Total Deaths in Asia (Highlight  1000)")
    plt.ylabel("Total Deaths")
    plt.xlabel("Disaster Type")

    plt.tight_layout()
    plt.show()
```

```python
visualize_asia_storm_earthquake(df)
```
